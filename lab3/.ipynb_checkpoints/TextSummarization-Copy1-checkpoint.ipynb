## Extraction based Text Summarization using NLTK


#Source: Stackabuse.com (Usman Malik)

```python
!pip install lxml
```

```python
import nltk
nltk.download('punkt')
```

#### Read web url 

```python
import bs4 as bs
import urllib.request
import re

scraped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Artificial_intelligence')
article = scraped_data.read()

parsed_article = bs.BeautifulSoup(article,'lxml')

paragraphs = parsed_article.find_all('p')

article_text = ""

for p in paragraphs:
    article_text += p.text
```

```python
print (article_text)
```

#### Read square brackets with numbers and extrac spaces

```python
# Removing Square Brackets and Extra Spaces
article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)
article_text = re.sub(r'\s+', ' ', article_text)
```

```python
print (article_text)
```

#### Remove special characters

```python
# Removing special characters and digits
formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text )
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
```

#### Change the above code to not to remove digits, ".", and ","

```python
formatted_article_text
```

```python
sentence_list = nltk.sent_tokenize(formatted_article_text)
```

```python
sentence_list
```

#### Calculate weighted frequency of words

```python
stopwords = nltk.corpus.stopwords.words('english')

word_frequencies = {}
for word in nltk.word_tokenize(formatted_article_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
```

```python
maximum_frequncy = max(word_frequencies.values())

for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
```

#### Score each sentence by adding weighted frequencies of words in the sentence

```python
sentence_scores = {}
for sent in sentence_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
```

#### Pick the sentences with most score

```python
import heapq
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = ' '.join(summary_sentences)
print(summary)
```

## Text Summarization using Gensim

```python
#!pip install beautifulsoup4
```

```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
import urllib, bs4
```

```python
scraped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Artificial_intelligence')
article = scraped_data.read()

parsed_article = bs4.BeautifulSoup(article,'lxml')

paragraphs = parsed_article.find_all('p')

article_text = ""

for p in paragraphs:
    article_text += p.text
```

```python
summarize(article_text, word_count=300)
```

```python
keywords(article_text, ratio=.01, lemmatize=True, scores=True)
```
